# Violin Plot

## Introduction

The violin plot is a versatile and informative visualization tool in the data scientist's arsenal.  By blending the strengths of box plots and kernel density estimation (KDE), violin plots offer a more nuanced look into the distribution of a dataset.

**Why Violin Plots?**

* **Beyond the Box:** While box plots are excellent for summarizing key statistics (median, quartiles, outliers), they can obscure the underlying shape of the distribution. Violin plots reveal this shape, showing where data is concentrated or sparse.
* **Density Insights:** The KDE component of a violin plot provides a smooth estimate of how data is spread out. This helps identify multi-modality (multiple peaks) and subtle asymmetries that box plots might miss.
* **Comparison Power:** Violin plots make it easy to compare distributions across different groups or categories within your data.

## Specify a color for each distribution

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')

# creating a dictionary with one specific color per group:
my_pal = {"versicolor": "g", "setosa": "b", "virginica": "m"}

# plot it
sns.violinplot(x=df["species"], y=df["sepal_length"], data=df, palette=my_pal, hue="species")
plt.show()
```

## Highlight a group


```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')

# creating a dictionary composed of species as keys and colors as values: red for the interesting group, blue for others
my_pal = {species: "r" if species ==
          "versicolor" else "b" for species in df["species"].unique()}

# make the plot
sns.violinplot(x=df["species"], y=df["sepal_length"], data=df, palette=my_pal, hue="species")
plt.show()
```

## Directly specyfing a list


```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
  
sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')

# specifying the group list as 'order' parameter and plotting
sns.violinplot(x='species', y='sepal_length', data=df, order=[ "versicolor", "virginica", "setosa"])
plt.show()
```

## Ordering by decreasing median


```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
  
sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')
 
# Using pandas methods and slicing to determine the order by decreasing median
my_order = df.groupby(by=["species"])["sepal_length"].median().sort_values().iloc[::-1].index
 
# Specifying the 'order' parameter with my_order and plotting
sns.violinplot(x='species', y='sepal_length', data=df, order=my_order)
plt.show()
```

## Grouped violin plot


```python
df = sns.load_dataset('tips')
sns.set_theme(style="darkgrid")
 
# Grouped violinplot
sns.violinplot(x="day", y="total_bill", hue="smoker", data=df, palette="Pastel1")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
 
# Grouped violinplot
sns.violinplot(x="smoker", y="total_bill", hue="day", data=df, palette="Pastel1")
plt.show()
```
