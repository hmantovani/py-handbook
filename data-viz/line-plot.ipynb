# Line Plot

A line plot is used to visualize data points connected by straight lines, typically over a continuous range of values. It's a valuable tool for displaying trends and patterns over time or across a continuous sequence. Line plots are commonly used in time series analysis and help reveal changes in data values and trends.

Common Use Cases:
- Visualizing time series data.
- Displaying trends and changes over a continuous range.
- Comparing data points across a continuous scale.

## The dataset

The FMRI (Functional Magnetic Resonance Imaging) dataset is a dataset provided by the Seaborn library, commonly used for demonstrating data visualization and analysis techniques. This dataset contains information related to brain activity measurements obtained through fMRI scans. Specifically, it includes data about various time points, the signal intensity recorded during these time points, and the type of event that occurred during the scans.

Researchers and data analysts often use the FMRI dataset to explore and visualize brain activity patterns in response to different events or stimuli. By examining the time-series data and signal intensity measurements, they can gain insights into how specific events or conditions influence brain activity, making it a valuable resource for studying neural responses and conducting data analysis and visualization exercises using Seaborn and other data analysis tools.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using the FMRI dataset from Seaborn
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event")
plt.title('Signal Intensity Line Plot')
plt.xlabel('Timepoint')
plt.ylabel('Signal Intensity')
plt.legend(title='Event')
plt.show()
```
