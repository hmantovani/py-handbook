# Pie Chart

A pie chart is a circular chart divided into sectors, each representing a proportion of a whole. It's used to show the composition of a dataset in terms of parts relative to the whole. Pie charts are helpful when you want to visualize the distribution of categorical data or compare parts to the total.

Common Use Cases:
- Displaying the composition of a whole in terms of its parts.
- Highlighting the relative proportions of different categories within a dataset.
- Showing the distribution of categorical data in a visually intuitive way.

## The dataset
The Titanic dataset is a widely recognized and frequently used dataset available through the Seaborn library. It is based on historical data from the infamous RMS Titanic voyage, which tragically sank in 1912 after hitting an iceberg. This dataset comprises information about passengers who were on board, including their demographics, ticket class, boarding location, and, notably, their survival status.

The Titanic dataset usually includes features such as passenger class, name, gender, age, number of siblings/spouses on board, number of parents/children on board, ticket fare, cabin number, and embarkation point. However, it is common to work with a simplified version that includes a subset of these features.

This dataset is often used for data analysis and machine learning tasks. Researchers and data analysts explore the Titanic dataset to uncover insights about passenger demographics, survival rates, and factors influencing survival. It serves as a valuable resource for illustrating data preprocessing, feature engineering, and predictive modeling techniques. Additionally, it carries a historical context that makes it engaging and relatable for both beginners and experienced practitioners in data science and analytics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using the Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
survived_count = titanic["survived"].value_counts()
plt.pie(survived_count, labels=["Did Not Survive", "Survived"], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Titanic Survival Pie Chart')
plt.show()
```
